## BASICS OF READING IN NETCDF FILE: USING XARRAY
- **Netcdf formatted data**: best to use the Xarray package - *much more to follow*
  - designed specifically to:
      - handle multi-D arrays (especially from Netcdf files), especially **spatial** data, including *labels*
      - efficiently do label-based operations on the arrays
<br><br>
- **two types of data structures you read into using Xarray:**
  - **DataArray**: a single multi-dimensional variable + coordinates
  - **DataSet**:  multiple variables that may or may not have same coordinates; multiple DataArrays
<br><br>
- **syntax**: *xr.open_dataarray(filename)* or *xr.open_dataset(filename)*
<br><br>

```python
import xarray as xr # load the xarray module
```

```python
# Read in the data - into a DataSet as it contains multiple variables
filename = 'pgbh00.gdas.1993031412.nc'
t = xr.open_dataset(filename) # Climate Forecast System Reanalysis data 
                                                 # from March 14, 1993 
```

```python
# List the variables in the DataSet
t.data_vars
```

```python
# The coordinates in the DataSet
t.coords
```

```python
# Access a specific variable in the DataSet
t['Pressure_reduced_to_MSL'] 
```

```python
# Print the values of the specific variable
t["Pressure_reduced_to_MSL"].values
```

- **Simple plotting using Xarray**
  - the plot method of DataArray calls plt.plot() from Matplotlib - is a "wrapper"

```python
t['Pressure_reduced_to_MSL'].plot()  # special plotting routine built into Xarray!
```
