## OBJECT-ORIENTED PLOTTING
- a Matplotlib plot has a *figure* object and at least one *axes* object
<br><br>
    - they are automatically produced by pyplot 
<br><br>
- to be able to better customize our plots (e.g., ability to have multiple axes if we'd like), we're going to explicitly create these objects
<br><br>
- **example**

```python
# If you're going in order of the lessons
# You have already downloaded this CSV data file

import matplotlib.pyplot as plt # function in matplotlib that produces quick and easy plots
import pandas as pd # load the Pandas module - much more to follow on Pandas!

data = pd.read_csv('CU_temp_data_Oct_14.csv', index_col='Day') # Oct 2014 temperature data from Champaign-Urbana 
```

```python
# One of the ways to create our fig and ax objects!
fig, ax = plt.subplots()
# Plot both high temperature and low temperature data
ax.plot(data['High'])
ax.plot(data['Low'])

# Note the changes in syntax
ax.set_title('Daily Temperatures in October 2014')
ax.set_xlabel('Day')
ax.set_ylabel('Temperature (\u00B0F)')
ax.legend(['High Temp','Low Temp']) # simple legend
```

## SAVING FIGURES
- **syntax**: fig.savefig
- do not call plt.show() prior as that resets the plot

```python
fig.savefig('Temperatures_Oct2014.png')
```
