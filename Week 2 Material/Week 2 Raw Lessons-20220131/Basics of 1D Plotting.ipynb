## MATPLOTLIB
- for our initial, simplest plotting, we'll use **pyplot**, a set of functions in Matplotlib that will give us Matlab-like functionality 
- *later*: plotting using figures and axes objects

## LINE PLOT
- let's start with the simplest **line plot** (more customizable plots later!)
- **syntax**: plt.plot(xdata,ydata)

```python
import matplotlib.pyplot as plt # function in matplotlib that produces quick and easy plots
import pandas as pd # load the Pandas module - much more to follow on Pandas!

data = pd.read_csv('CU_temp_data_Oct_14.csv', index_col='Day') # Oct 2014 temperature data from Champaign-Urbana 
```

```python
# Good quick and easy plot, but can't customize everything

plt.plot(data.High) # Plotting maximum temperature over the month 
plt.plot(data.Low) # Let's also plot minimum temperature!
plt.xlabel('Day') # x-axis label
plt.ylabel('Daily Temperature (\u00B0F)') # y-axis label
plt.title('Daily Temperatures in October 2014') # plot title
plt.legend(['High Temp','Low Temp']) # simple legend

plt.show() # Displays all open figures
           # No more changes can be made after calling this 
```

## SCATTER PLOT
- we use **scatter plots** to examine statistical relationships (like correlations) between two variables
- **syntax**: scatter(xdata,ydata)

```python
plt.scatter(data.High,data.Low) # can customize size, color, type of marker as well 
plt.xlabel('High Temperature (\u00b0F)') # degree symbol...more on that later!
plt.ylabel('Low Temperature (\u00b0F)') 
plt.title('Daily Temperatures in October 2014')
```

## BOX AND WHISKER PLOT
- we use **box and whisker plots** to examine characteristics of the data and its distribution
- **syntax**: boxplot(data)

```python
plt.boxplot(data.High)
plt.title('High Temperatures')
```

##  HISTOGRAM
- we use **histograms** to examine the statistical distribution of data
- **syntax**: hist(data,nbins)

```python
data2 = pd.read_csv('earthquake_data_USA.csv') # Dec 2019-Dec 2020 US Earthquake Data > 2.5  
plt.hist(data2.mag,10) # data, number of bins 
plt.xlabel('Earthquake Magnitude')
plt.ylabel('Number of Events')
plt.title('Earthquake Events > Magnitude 2.5, Dec 2019-2020')
```
